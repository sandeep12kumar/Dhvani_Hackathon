```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp


def lorenz(t, state, a, b, c):
    x, y, z = state
    dxdt = a * (y - b)
    dydt = b * x - y - x * z
    dzdt = x * y - c * z
    return [dxdt, dydt, dzdt]


a = 10
b = 28
c = 2.667


x0 = 0
y0 = 1
z0 = 1.05
initial_state = [x0, y0, z0]


t_span = (0, 40)
t_eval = np.linspace(t_span[0], t_span[1], 10000)


solution = solve_ivp(lorenz, t_span, initial_state, t_eval=t_eval, args=(a, b, c))


fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(solution.y[0], solution.y[1], solution.y[2], lw=0.5, color='b')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Lorenz System Trajectory (Bee Path)')


plt.show()
```
